##  TOKENIZATION

```python
from nltk.tokenize import word_tokenize
dataset['text']=dataset['text'].apply(word_tokenize)
```

```python
dataset
```

```python
print(dataset.head(10))
```

## STEMMING 

```python
from nltk.stem.snowball import SnowballStemmer
porter = SnowballStemmer("english", ignore_stopwords=False)
```

```python
def stem_it(text):
    return [porter.stem(word) for word in text]
```

```python
dataset['text']=dataset['text'].apply(stem_it)
```

```python
dataset
```

```python
print(dataset.head(10))
```

## Stopword removal

```python
def stop_it(text):
    return [word for word in text if len(word) >> 2] 
```

```python
dataset['text']=dataset['text'].apply(stop_it)
```

```python
dataset
```

```python
print(dataset['text'].head(10))
```

```python
dataset['text']=dataset['text'].apply(' '.join)
```

```python
dataset
```

## Splitting up of data 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset['text'], dataset['genuiness'], test_size=0.25)
```

```python
x_train
```

## Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
my_tfidf = TfidfVectorizer( max_df=0.7)

tfidf_train = my_tfidf.fit_transform(x_train)
tfidf_test = my_tfidf.transform(x_test)
```

```python
print(tfidf_train)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_1 = LogisticRegression(max_iter=500)
model_1.fit(tfidf_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
pred_1 = model_1.predict(tfidf_test)
cr1    = accuracy_score(y_test,pred_1)
print(cr1*100)
```

## Passive Aggressive Classifier

```python
from sklearn.linear_model import PassiveAggressiveClassifier

model = PassiveAggressiveClassifier(max_iter=100)
model.fit(tfidf_train, y_train)
```

```python
y_pred = model.predict(tfidf_test)
accscore = accuracy_score(y_test, y_pred)
print('The accuracy of prediction is ',accscore*100)
```
